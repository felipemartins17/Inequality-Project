```python
import matplotlib
import decimal as Decimal
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# All data used in this analysis were got from https://databank.worldbank.org/

# Import the World's GDP indicators 
gdp_world = pd.read_csv('GDP_World.csv')
```

```python
gdp_world.head()
```

```python
# We want to look further into the G20 group which represents 80% of World's GDP. G20 is made of 19 countries + Europe Union. 
# We're not considering Europe Union in our analysis, let's focus in the countries

g20s = ['Argentina', 'Australia', 'Brazil', 'Canada', 'China', 'France', 'Germany', 'Indonesia', 'India','Italy', 'Japan',
'Mexico', 'Korea, Rep.', 'Russian Federation', 'South Africa', 'Saudi Arabia', 'Turkey', 'United Kingdom', 'United States']
gdp_g20 = gdp_world[gdp_world['Country Name'].isin(g20s)]
```

```python
gdp_g20.head(5)
```

```python
# Now we got the G20 GDP by country, lets plot it and take a look

# Before plot, we need to convert our main column to numeric, I'll take 2014 for this:

gdp_g20['2014'] = gdp_g20['2014'].apply(lambda x: x.replace('$', '').replace(',', ''))
gdp_g20['2014'] = pd.to_numeric(gdp_g20['2014'])
```

```python
ax = gdp_g20.plot.bar(x='Country Name', y='2014', rot=0)

ax.set_xlabel('G20 Countries')
ax.set_ylabel('GDP - (US$)- Trillions')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Our main intention is to see how GDP and Social Inequality impacts violence, so let's now look at the Gini index of g20's countries

gini_world = pd.read_csv('gini_idx.csv')
gini_g20 = gini_world[gini_world['Country Name'].isin(g20s)]
gini_g20.head(20)
```

```python
# We can observe that there are many null values in the DF, we also can notice that 2014 is the year with the most information 
# available, so let's take a look on it

ax = gini_g20.plot.bar(x='Country Name', y='2014', rot=0)

ax.set_xlabel('G20 Countries')
ax.set_ylabel('Gini - Desigualdade')
plt.xticks(rotation = 90)
plt.show()
```

```python
# So now we got a problem to solve. Gini information is missing for India, Saudi Arabia and Japan
# I checked the world https://data.worldbank.org/ and see that Saudi Arabia never provided this index, so my only options is take 
# Saudi Arabia out of the list
# For Japan, the most recent Gini index is from 2013. As it's unlikely to the index change much from one year to another. 
# I'll input the value manually, which is 32.9
# India has provided info in for 3 times in this century, I'm going to use the mean value of it, see below:

gini_india = pd.read_csv('India_gini.csv')
gini_india = gini_india.dropna(axis='columns')
gini_india.head()
gini_india['mean'] = gini_india.mean(axis=1)
gini_india.head()
```

```python
# So now let's input what we found out and also drop Saudi Arabia

gini_g20.loc[gini_g20['Country Name'] == 'India', "2014"] = 35.17
gini_g20.loc[gini_g20['Country Name'] == 'Japan', "2014"] = 32.9

indexNames = gini_g20[gini_g20['Country Name'] == 'Saudi Arabia'].index
gini_g20.drop(indexNames,inplace=True)
gini_g20.head(20)

```

```python
# Lets plot again and see how it looks like

ax = gini_g20.plot.bar(x='Country Name', y='2014', rot=0)

ax.set_xlabel('G20 Countries')
ax.set_ylabel('Gini - Desigualdade')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Now its time to look at the homicides index 

homi_world = pd.read_csv('homicides_world.csv')
homi_g20 = homi_world[homi_world['Country Name'].isin(g20s)]
homi_g20.head(20)
```

```python
#Lets drop Saudi Arabia and see what we find about Turkey

indexNames = homi_g20[homi_g20['Country Name'] == 'Saudi Arabia'].index
homi_g20.drop(indexNames,inplace=True)
homi_g20.head(20)
```

```python
# Lets take a look on Turkey's homicides index in this century

homi_turkey = pd.read_csv('homi_turkey.csv')
```

```python
homi_turkey.head(20)
```

```python
#Its possible to see how the homicides index decreased from 2003 to 2015. So I choose to consider the mean of years availabel 
#from 2010 to 2018

homi_turkey_up = homi_turkey[['Country Name','Country Code', '2010', '2011', '2012', '2015', '2016', '2017', '2018']]
homi_turkey_up.head()

homi_turkey_up['mean'] = homi_turkey_up.mean(axis=1)
homi_turkey_up.head()
```

```python
#Now lets input what we found out

homi_g20.loc[homi_g20['Country Name'] == 'Turkey', "2014"] = 3.51041
```

```python
homi_g20.head(20)
```

```python
ax = homi_g20.plot.bar(x='Country Name', y='2014', rot=0)

ax.set_xlabel('G20 Countries')
ax.set_ylabel('Homicides per 100 people')
plt.xticks(rotation = 90)
plt.show()
```

```python
# We also want to see how poverty impacts violence

pov_world = pd.read_csv('world_poverty.csv')
pov_g20 = pov_world[pov_world['Country Name'].isin(g20s)]
pov_g20.head(20)
```

```python
#The only index available for Japan is 0.7 in 2013, let's use it
#Lets look what India provided

india_pov = pd.read_csv('India_pov.csv')
india_pov.head()
```

```python
india_pov['mean'] = india_pov.mean(axis=1)
india_pov.head()
```

```python
pov_g20.loc[pov_g20['Country Name'] == 'India', "2014"] = 26.0
pov_g20.loc[pov_g20['Country Name'] == 'Japan', "2014"] = 0.7

```

```python
pov_g20.head(20)
```

```python
indexNames = pov_g20[pov_g20['Country Name'] == 'Saudi Arabia'].index
pov_g20.drop(indexNames,inplace=True)
pov_g20.head(20)
```

```python
#Lets plot and take a look

ax = pov_g20.plot.bar(x='Country Name', y='2014', rot=0)

ax.set_xlabel('G20 Countries')
ax.set_ylabel('Poverty')
plt.xticks(rotation = 90)
plt.show()
```

```python
gdp_g20_2014 = gdp_g20[['Country Name', 'Indicator Name', '2014']]
gdp_g20_2014.columns = ['Country Name', 'Indicator Name', 'GDP Value']
gdp_g20_2014.head()
```

```python
tempmerge = pd.merge(gdp_g20_2014, gini_g20['2014'], left_index=True, right_index=True)
tempmerge.columns = ['Country Name', 'Indicator Name', 'GDP Value', 'Gini Index']
tempmerge.head()

tempmerge = pd.merge(tempmerge, homi_g20['2014'], left_index=True, right_index=True)
tempmerge.columns = ['Country Name', 'Indicator Name', 'GDP Value', 'Gini Index', 'Homicides per 100 people']
tempmerge.head()

final_df = pd.merge(tempmerge, pov_g20['2014'], left_index=True, right_index=True)
final_df.columns = ['Country Name', 'Indicator Name', 'GDP Value', 'Gini Index', 'Homicides per 100 people', 'Poverty gap at $3.20 a day']
final_df.head()
```

```python
# Time to do some analysis 

corr = final_df.corr()
sns.heatmap(corr)
```

```python
# Here we're trying to see a correlaction of cause and consequence. Where GDP and Gini Index are the possible "causes" and 
# Homicides and Poverty the possible consequences
# So, it's possible to see in this correlaction heatmap that there's a posivite correlation between Gini Index and Homicides/
# /Poverty. Meaning that as higher is the social inequality in a country more homicides/poverty it tends to have
# Also, we can see a negative correlation between GDP and Homicides/Poverty. Meaning that as higher the GDP of a country is, 
# lower tends to be its homicides and poverty index


# Now let's look the top 7 countries in these indexes and see if we find out something interesting

#Top 7 most inequal countries

final_gini = final_df.nlargest(7, 'Gini Index')
final_gini.head(7)
```

```python
# Top 7 countries with more homicides

final_homi = final_df.nlargest(7, 'Homicides per 100 people')
final_homi.head(7)
```

```python
# Top 7 richest countries

final_gdp = final_df.nlargest(7, 'GDP Value')
final_gdp.head(7)
```

```python
# Top 7 countries with more poverty

final_pov = final_df.nlargest(7, 'Poverty gap at $3.20 a day')
final_pov.head(7)
```

```python
# Well, we can see that Brazil is the only country that figures in all 4 lists. Which seems like an exception for the 
# correlation we saw previously between GDP - homicides/poverty (since Brazil is the 7th country in GDP's list but still with
# high rates of Poverty, Homicides and Inequality
# It makes me curious of what index (GDP or Gini) as more impact in homicides and/or poverty 
# Let's plot these top7 to have a better view

fig, ax = plt.subplots() # Create matplotlib figure
width = 0.3

ax.bar(final_gdp['Country Name'], final_gdp['GDP Value'], color='blue', width=width)
ax.set_ylabel('GDP Value')
plt.xticks(rotation = 90)


ax2 = ax.twinx()
ax2.bar(final_gdp['Country Name'], final_gdp['Homicides per 100 people'], color='red', width=width, align='edge')
ax2.set_ylabel('Homicides per 100 people')

plt.show()

```

```python
fig, ax = plt.subplots() # Create matplotlib figure
width = 0.3

ax.bar(final_gini['Country Name'], final_gini['Gini Index'], color='blue', width=width)
ax.set_ylabel('Gini Index')
plt.xticks(rotation = 90)


ax2 = ax.twinx()
ax2.bar(final_gini['Country Name'], final_gini['Homicides per 100 people'], color='red', width=width, align='edge')
ax2.set_ylabel('Homicides per 100 people')

plt.show()
```
